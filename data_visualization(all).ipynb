```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
iris=sns.load_dataset ("iris")
```

```python
sns.lineplot(data=iris,x="sepal_length", y= "petal_length")
```

```python
sns.lineplot(data=iris,x="sepal_length", y= "petal_length",hue='species')
```

```python
sns.regplot(data=iris,x="sepal_length", y= "petal_length")
```

```python
sns.jointplot(data=iris,x="sepal_length", y= "petal_length")
```

```python
sns.pairplot(data=iris)
```

heatmap

```python
corr= iris.corr()
```

```python
sns.heatmap(corr)
```

```python
sns.boxplot(data=iris)
```
